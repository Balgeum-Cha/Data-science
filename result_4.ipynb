```python
import numpy as np
import pandas as pd
```

```python
# 데이터 불러오기
df = pd.read_excel('공개데이터_18.xlsx')
```

```python
lunfee = ['regname']
for i in range(1,6):
    lunfee.append(f'lunfee_{i}')
print(lunfee)
```

```python
lunfee_df = df[lunfee]
```

```python
local_lunfee = lunfee_df.groupby('regname').sum(0)
```

```python
local_lunfee.columns = ['교육청지원금액','자치단체지원금액','학교자체지원금액','보호자지원금액','기타지원금액']

local_lunfee
```

```python
print(local_lunfee.mean())
 
print(local_lunfee.var())
 
print(local_lunfee.std())
```
