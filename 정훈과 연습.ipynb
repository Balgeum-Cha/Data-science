```python
import numpy as np
import pandas as pd
```

```python
myfile = pd.read_excel('공개데이터_18.xlsx')
```

```python
seoul = myfile[myfile['regname'] == '서울']
```

```python
# 각 학교의 학년 여학생 남학생 변수 리스트
# 각 학생 수의 추이가 어떻게 변해가는지?
colnames = []
for i in range(1,7):
    colnames.append(f'numcls_m{i}')
    colnames.append(f'numcls_f{i}')
```

```python
num_total_students = seoul[colnames].sum(1)
```
