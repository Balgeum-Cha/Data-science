```python
def cleaner(df, year):https://docs.google.com/spreadsheets/d/1UPqEx62i8jR6Eam-xAPA4OMDpQ0yKUmegbLt3w7ASmM/edit#gid=1326651727
    df = df.reset_index()
    df = df.rename(columns={'regname': '지역'})
    df.insert(0, '년도', [year]*16)
    return df
```

```python
local_num_df15_sex = cleaner(local_num_df15_sex, 2015)
```

```python
local_num_df16_sex = cleaner(local_num_df16_sex, 2016)
```

```python
local_num_df17_sex = cleaner(local_num_df17_sex, 2017)
```

```python
local_num_df18_sex = cleaner(local_num_df18_sex, 2018)
```

```python
local_num_df_whole =  pd.concat([local_num_df15_sex,local_num_df16_sex,local_num_df17_sex,local_num_df18_sex],axis =0)
```

```python
local_num_df_whole.지역.unique().tolist()
```

```python
for idx, regname in enumerate(local_num_df_whole.지역.unique().tolist()):
    plt.figure(idx)
    temp_df = local_num_df_whole[local_num_df_whole['지역'] == regname ]
    ax = temp_df.plot(x='년도', y=['남자관리자', '여자관리자'], kind='bar',rot = 0)
    plt.title(regname, fontsize = 20)
    plt.xlabel('년도')
    plt.ylabel('관리자수(명)')
    values = temp_df['남자관리자'].values.tolist() + temp_df['여자관리자'].values.tolist()
    for i, p in enumerate(ax.patches):
        left, bottom, width, height = p.get_bbox().bounds
        ax.annotate(values[i], (left+width/2, height*1.01), ha='center')
```
