```python
import numpy as np
import pandas as pd
```

```python
# 데이터 불러오기
df = pd.read_excel('공개데이터_18.xlsx')
```

```python
# 학생 수 관련 데이터 선택
numcls = df.loc[:,'regname':'numcls_f6']
```

```python
# 관련없는 데이터 삭제
numcls = numcls.drop(['schlevel','schtypename','schgenname','schprin'],axis = 1)
```

```python
numcls.head()
```

```python
#데이터 정제 (비어 있는 값에 0 넣기)
list = []
for i in range(1,7):
    list.append(f'numcls_{i}')
    list.append(f'numcls_m{i}')
    list.append(f'numcls_f{i}')
print(list)

columns = list[:]
for ind in numcls.index:
    for col in columns:
        if type(numcls[col][ind]) != int:
            numcls[col][ind] = 0
```

```python
numcls
```

```python
# 학교 별 총 학급 수 구하기
num_class = []
for i in range(1,7):
    num_class.append(f'numcls_{i}')
num_class

numcls['학급수'] = numcls[num_class].sum(1)
numcls
```

```python
# 원하는 열 선택하려고 리스트 만듬
list2 = ['regname','학급수']
for i in range(1,7):
    list2.append(f'numcls_m{i}')
    list2.append(f'numcls_f{i}')
print(list2)

numcls_class = numcls[list2]
numcls_class
```

```python
# 지역별로 묶고 행끼리 더함
num_class_local = numcls_class.groupby('regname').sum(0)
num_class_local.head()
```

```python
# 열끼리 더해서 새로운 column 만듬
num_class_local['학생수'] = num_class_local.iloc[:,2:].sum(axis=1)
```

```python
# 학급수 학생수 데이터프래임 만들기
local_school = num_class_local[['학급수','학생수']]
local_school
```

```python
print('평균:')
print(local_school.mean())
print('분산: ')
print(local_school.var())
print('표준편차:')
print(local_school.std())
```

```python
# 지역 별 반 평균 학생수 구하기
num_class_local_student = num_class_local['학생수']/num_class_local['학급수']
```

```python
# 리스트로 만들기
num_students = num_class_local_student.tolist()
```

```python
# 리스트로 만들기
local = num_class_local_student.index.tolist()
```

```python
type(num_class_local_student)
```

```python
# 그래프 그리기
import matplotlib.pyplot as plt
plt.bar(local, num_students)
plt.title('지역별 반 평균 학생수')
plt.xlabel('지역', color = 'red')
plt.ylabel('학생수(명)',color = 'red')
for i,v in enumerate(local):
    plt.text(
        v,
        num_students[i],
        '{:.1f}'.format(num_students[i]),
        fontsize = 9,
        color = 'black',
        horizontalalignment='center',
        verticalalignment='bottom'
    )
plt.ylim(10,25)
```
