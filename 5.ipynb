```python
# 원하는 행 리스트로 가져오기
list2 = ['regname']
for i in range(1,7):
    list2.append(f'numcls_m{i}')
    list2.append(f'numcls_f{i}')
print(list2)

num_df18 = df18[list2]
num_df16 = df16[list2]
num_df15 = df15[list2]
num_df17 = df17[list2]
```

```python
# 반복문 쓰기 위해서 리스트 만듬, 데이터프레임에 데이터들 넣음
dataframes = [num_df15, num_df16, num_df17, num_df18]
list = []
for i in range(1,7):
    list.append(f'numcls_m{i}')
    list.append(f'numcls_f{i}')
print(list)

columns = list[:]
```

```python
# int 가 아닌 값 0으로 변환
for dataframe in dataframes:
    for idx in dataframe.index:
        for col in columns:
            if type(dataframe[col][idx]) != int:
                dataframe[col][idx] = 0
```
